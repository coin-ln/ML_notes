# 一元线性回归梯度下降法

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
datas=np.genfromtxt("data.csv",delimiter=',')
datas
```

```python
X=datas[:,0]
Y=datas[:,1]
X,Y
plt.scatter(X,Y)
plt.show()
```

```python
def get_lost(k,b,X,Y):
    res=0.0
    for i in range(len(X)):
        res=res+(k*X[i]+b-Y[i])**2
    return res/2.0/len(X)
# test1,test2=get_lost(1,1,[1],[1]),get_lost(1,1,[1],[2])
# print(test1,test2)
```

```python
def gradient_descent(X,Y,k,b,lr,times):
    m=len(X)
    for i in range(times):
        k_grad=0.0
        b_grad=0.0
        for j in range(m):
            k_grad+=(1/m)*X[j]*(k*X[j]+b-Y[j])
            b_grad+=(1/m)*(k*X[j]+b-Y[j])
        k-=k_grad*lr
        b-=b_grad*lr
        
        # if i%10==0:
        #     print("times:",i)
        #     plt.plot(X,Y,'b.')
        #     plt.plot(X,k*X+b,'r') 
        #     plt.show()
    return k,b
```

```python
#初始值
# k=0.0       #斜率
# b=0.0       #截距
lr=0.1     #学习率
times=50     #迭代次数
for i in range(10):
    k=1.0
    b=0.0
    lr/=10
    k,b=gradient_descent(X,Y,k,b,lr,times)
    print("初始值为b={0},k={1},lr={2}".format(b,k,lr))
    print("迭代{0}次后,损失函数的值为{1}".format(times,get_lost(k,b,X,Y)))
    plt.plot(X,Y,'b.')
    plt.plot(X,k*X+b,'r') 
    plt.show()

```
